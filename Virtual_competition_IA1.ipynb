````python
sugar=pd.read_csv('/content/diabetes_model.csv')
sugar
````

# Univariate Analysis

**Histograms: To understand the distribution of each variable.**

````python
sns.histplot(sugar['BMI'], bins=30, edgecolor='black',color='slateblue')
plt.show()
````

The distribution is slightly skewed to the right, meaning there are a few individuals with higher BMI values that extend the tail on the right side.
The central tendency is around 30-35 which means that there are more no:of people with the BMI of that range.

````python
plt.figure(figsize=(8,6))
sns.boxplot(data=sugar[['Age']],width=.3)
plt.xticks(rotation=90)  # Rotate x labels for better readability
plt.show()
````

There is some outlier on the right side, indicating an individual whose age is significantly higher than the majority of the sample.

# Bivariate Analysis:

**Plot scatter plots for pairs of variables**

````python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Glucose', y='Insulin', data=sugar, color='black')
plt.title('Glucose vs Insulin')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
````

**Positive Correlation:**  This means that as glucose levels increase, insulin levels tend to increase as well.

**Outliers:** There are a few outliers

````python
# Compute correlation matrix
sugar_data=sugar[['BMI','BloodPressure']]
correlation_matrix = sugar_data.corr()

# Plot correlation matrix
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='seismic', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
````

 The correlation coefficient between BMI and Blood Pressure is 0.28. This indicates a weak positive correlation between them. In other words, as BMI increases, there's a slight tendency for Blood Pressure to increase as well, but the relationship is not very strong.

# Multivariate Analysis:

**Pair Plots: To visualize pairwise relationships in the dataset.**

````python
pair_columns = ['Glucose', 'Insulin', 'Pregnancies','Outcome']
pair_plot = sugar[pair_columns]

sns.pairplot(pair_plot,hue='Outcome')
plt.show()
````

 There appears to be a positive correlation between "Glucose" and "Insulin".This suggests that individuals with higher glucose levels also tend to have higher insulin levels.


 The density plots for "Pregnancies" in the diagonal plots reveal that individuals with diabetes (Outcome=1) tend to have a slightly higher average number of pregnancies compared to those without diabetes (Outcome=0).

Glucose" and "Insulin" have a skewed distribution, with a tail extending to the right, indicating the presence of some outliers with very high values.

**3D Plots: To explore interactions among three variables.**

````python
from mpl_toolkits.mplot3d import Axes3D

x = sugar['BMI']
y = sugar['Insulin']
z = sugar['Age']

fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z,color='Purple', marker='o')
ax.set_xlabel('BMI')
ax.set_ylabel('Insulin')
ax.set_zlabel('Age')
plt.title('3D Scatter Plot of Glucose, Insulin, and BMI')
plt.show()
````

Clustering: There is clustering of data points in the lower regions of the plot, suggesting a concentration of individuals with lower BMI, glucose, and insulin levels.
Scatter: The data points are somewhat scattered, indicating individual variability in the relationships between the three variables.
Outliers: There are a few outliers, especially in the higher regions of the plot, representing individuals with higher BMI, glucose, and insulin levels.
